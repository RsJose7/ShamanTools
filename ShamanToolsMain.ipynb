```python
# SISTEMA DE INVENTARIOS DE SHAMAN TOOLS

# Variables del sistema
usuarios = [
    {"nombreUsuario": "Ren Tao", "contraseña": "1234_4321", "rol": "admin"},
    {"nombreUsuario": "Ioh Asakura", "contraseña": "abc_cba", "rol": "vendedor"},
    {"nombreUsuario": "Chocolove McDonnell", "contraseña": "11110000", "rol": "vendedor"}
]

productos = [
    {"codigo": "001", "nombreProducto": "Ladrillo 10x20x30", "precioCompra": 1200, "precioVenta": 1500, "cantidad": 1000},
    {"codigo": "002", "nombreProducto": "Cemento Gris x50kg", "precioCompra": 28000, "precioVenta": 30000, "cantidad": 200},
    {"codigo": "003", "nombreProducto": "Varilla Corrugada #5 x6m", "precioCompra": 33000, "precioVenta": 35000, "cantidad": 150},
    {"codigo": "004", "nombreProducto": "Impermeabilizante 22kg", "precioCompra": 350000, "precioVenta": 360000, "cantidad": 50},
    {"codigo": "005", "nombreProducto": "Malla Electro Soldada 6x2.2m", "precioCompra": 42000, "precioVenta": 45000, "cantidad": 75}
]
# Registro de ventas comenzado en 0
ventas = []


# Función de inicio de sesión
def iniciarSesion():
    nombreUsuario = input("\nIngrese su nombre de usuario: ")
    contraseña = input("Ingrese su contraseña: ")

    for usuario in usuarios:
        if usuario["nombreUsuario"] == nombreUsuario and usuario["contraseña"] == contraseña:
            print("Inicio de sesión exitoso")
            return usuario["rol"]
    print("Error de autenticación")
    return


# Función para registrar un nuevo usuario (para admin)
def registrarUsuario():
    nombreUsuario = input("Ingresar el nuevo nombre de usuario: ")

    for usuario in usuarios:
        if usuario["nombreUsuario"] == nombreUsuario:
            print("El usuario ya está registrado")
            return

    contraseña = input("Ingresa una contraseña: ")
    rol = input("Ingrese el rol del usuario (admin/vendedor): ")
    nuevoUsuario = {"nombreUsuario": nombreUsuario, "contraseña": contraseña, "rol": rol}
    usuarios.append(nuevoUsuario)
    print("Usuario registrado exitosamente")


# Función para agregar o actualizar productos (para admin)
def agregarActualizarProducto():
    codigo = input("Ingresa el código del producto: ")
    nombreProducto = input("Ingresa el nombre del producto: ")
    precioCompra = float(input("Ingresa el precio de compra del producto: "))
    precioVenta = float(input("Ingresa el precio de venta del producto: "))
    cantidad = int(input("Ingresa la cantidad: "))

    encontrado = False
    for producto in productos:
        if producto["codigo"] == codigo:
            producto["cantidad"] += cantidad
            producto["precioCompra"] = precioCompra
            producto["precioVenta"] = precioVenta
            encontrado = True
            print("Producto actualizado exitosamente")
            break
    if not encontrado:
        nuevoProducto = {"codigo": codigo, "nombreProducto": nombreProducto, "precioCompra": precioCompra, "precioVenta": precioVenta, "cantidad": cantidad}
        productos.append(nuevoProducto)
        print("Producto agregado exitosamente")


# Función para consultar productos
def consultarProductos():
    opcion = input("Elige una opción (lista/buscar): ")

    if opcion == "lista":
        print("Lista de productos:")
        for producto in productos:
            print(producto)
    elif opcion == "buscar":
        codigo = input("Ingresa el código del producto: ")
        for producto in productos:
            if producto["codigo"] == codigo:
                print("Producto:", producto)
                return
        print("Producto no encontrado")


# Función para registrar ventas (para vendedor)
def registrarVenta():
    codigo = input("Ingresa el código del producto: ")
    cantidadVendida = int(input("Ingresa la cantidad vendida: "))

    for producto in productos:
        if producto["codigo"] == codigo:
            if producto["cantidad"] >= cantidadVendida:
                producto["cantidad"] -= cantidadVendida
                venta = {"codigo": codigo, "cantidadVendida": cantidadVendida, "total": producto["precioVenta"] * cantidadVendida}
                ventas.append(venta)
                print("Venta registrada exitosamente")
            else:
                print("Cantidad insuficiente en inventario")
            return
    print("Producto no encontrado")


# Función para comprar productos (para admin)
def comprarProducto():
    codigo = input("Ingresa el código del producto: ")
    cantidadNueva = int(input("Ingresa la cantidad a agregar en el inventario de productos: "))

    encontrado = False
    for producto in productos:
        if producto["codigo"] == codigo:
            producto["cantidad"] += cantidadNueva
            encontrado = True
            print("Cantidad actualizada en el inventario")
            break
    if not encontrado:
        nombreProducto = input("Ingresa el nombre del producto: ")
        precioCompra = float(input("Ingresa el precio de compra: "))
        precioVenta = float(input("Ingrese el precio de venta: "))
        nuevoProducto = {"codigo": codigo, "nombreProducto": nombreProducto, "precioCompra": precioCompra, "precioVenta": precioVenta, "cantidad": cantidadNueva}
        productos.append(nuevoProducto)
        print("Nuevo producto agregado exitosamente")


# Función para ver resumen de ventas (para admin)
def verResumenVentas():
    print("Resumen de ventas:")
    for venta in ventas:
        print(venta)


# Estructura principal del sistema
def sistemaShamanTools():
    print("\nBienvenido a Shaman Tools")
    print("Seleccione una opción:")
    print("1. Iniciar sesión")
    print("2. Salir")
    opcion = input("Ingrese una opción: ")
    if opcion == "1":
        rol = iniciarSesion()
    if opcion == "2":
        print("Sesion Cerrada")
        return

    if rol == "admin":
        while True:
            print("\nBienvenido a Shaman Tools")
            print("Seleccione una opción:")
            print("1. Registrar usuario")
            print("2. Agregar o actualizar producto")
            print("3. Consultar productos")
            print("4. Comprar productos")
            print("5. Ver resumen de ventas")
            print("6. Salir")
            opcion = input("Ingrese una opción: ")

            if opcion == "1":
                registrarUsuario()
            elif opcion == "2":
                agregarActualizarProducto()
            elif opcion == "3":
                consultarProductos()
            elif opcion == "4":
                comprarProducto()
            elif opcion == "5":
                verResumenVentas()
            elif opcion == "6":
                print("Sesion Cerrada")
                return sistemaShamanTools()
            continuar = input("¿Desea realizar otra operación? (sí/no): ")
            if continuar == "sí":
                return opcion
            else:
                print("Sesion Cerrada")
                return sistemaShamanTools()
    elif rol == "vendedor":
        while True:
            print("\nBienvenido a Shaman Tools")
            print("Seleccione una opción:")
            print("1. Consultar productos")
            print("2. Vender productos")
            print("3. Salir")
            opcion = input("Ingrese una opción: ")

            if opcion == "1":
                consultarProductos()
            elif opcion == "2":
                registrarVenta()
            elif opcion == "3":
                print("Sesion Cerrada")
                return sistemaShamanTools()


            continuar = input("¿Desea realizar otra operación? (sí/no): ")
            if continuar == "sí":
                return opcion
            else:
                print("Sesion Cerrada")
                return sistemaShamanTools()
    else:
        print("Credenciales inválidas. Sesión cerrada")


# Volver a iniciar el sistema
sistemaShamanTools()
```
